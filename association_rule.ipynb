```python
# Data Preprocessing:
import pandas as pd
df = pd.read_excel('Market_data.xlsx')
df
```

```python
# Slice Column

columns_to_slice = ['BillNo', 'Itemname']
sliced_df = df[columns_to_slice]
sliced_df.head(10)
```

```python
columns_to_slice = ['BillNo', 'Itemname']
sliced_df = df[columns_to_slice]

# Create a dictionary to store items for each BillNo
bill_items = {}

# Process the data and populate the dictionary
for index, row in sliced_df.iterrows():
    BillNo = row['BillNo']
    Itemname = row['Itemname']
    
    if BillNo in bill_items:
        bill_items[BillNo].append(Itemname)
    else:
        bill_items[BillNo] = [Itemname]

# Convert the dictionary values to sets
result = {BillNo: set(Itemname) for BillNo, Itemname in bill_items.items()}

# Print the result
for BillNo, Itemname in result.items():
    print(f"{BillNo}\t\t{Itemname}")
```

```python
# Create a new DataFrame from the result
result_df = pd.DataFrame(list(result.items()), columns=['BillNo', 'Itemname'])

# Print the result DataFrame
print(result_df)
```

```python
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.preprocessing import TransactionEncoder

# Convert the 'Itemname' column to a list of lists
transactions = result_df['Itemname'].apply(list).tolist()

# Use TransactionEncoder to one-hot encode the transaction data
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)

# Convert the one-hot encoded array to a DataFrame
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

# Apply FP-Growth algorithm
frequent_itemsets = fpgrowth(df_encoded, min_support=0.2, use_colnames=True)

# Print the frequent item sets
print(frequent_itemsets)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Assuming your dataset is stored in 'result_df'
transactions = result_df['Itemname'].tolist()

# Use TransactionEncoder to one-hot encode the transactions
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)

# Convert the one-hot encoded transactions back to a DataFrame
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

# Apply Apriori algorithm
frequent_itemsets = apriori(df_encoded, min_support=0.3, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Print the results
print("Frequent Itemsets:")
print(frequent_itemsets)
print("\nAssociation Rules:")
print(rules)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MultiLabelBinarizer

# Assuming your dataset is stored in 'result_df'
sequences = result_df['Itemname'].tolist()

# Use MultiLabelBinarizer to create a binary matrix
mlb = MultiLabelBinarizer()
binary_matrix = pd.DataFrame(mlb.fit_transform(sequences), columns=mlb.classes_)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
cluster_labels = kmeans.fit_predict(binary_matrix)

# Add cluster labels to the original dataset
result_df['Cluster'] = cluster_labels

# Print the results
print("Clustered Dataset:")
print(result_df)
```
